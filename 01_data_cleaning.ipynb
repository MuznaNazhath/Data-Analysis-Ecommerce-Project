```python
import pandas as pd

# Read the CSV file
df = pd.read_csv(r'C:/Users/muzna/anaconda_projects/ecommerce_sample.csv', parse_dates=['InvoiceDate'])

df.head()
```

```python
# Basic cleaning
df = df.drop_duplicates()
df = df[df['Quantity'] != 0]  # remove returns if treating separately
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
df['CustomerID'] = df['CustomerID'].fillna('Unknown')

df.head()
```

```python
# Normalize text
df['Description'] = df['Description'].str.strip().str.title()

df.head()
```

```python
# Export cleaned - fixed path to save in the same directory as the original file
df.to_csv(r'C:/Users/muzna/anaconda_projects/ecommerce_cleaned.csv', index=False)
```
